```python
vacation1_df = city_data_df.loc[city_data_df["City"]=="Silleda", ["City", "Country", "Max Temp", "Current Description", "Lat", "Lng", "Hotel Name"]]
```

```python
vacation2_df = city_data_df.loc[city_data_df["City"]=="Isernia", ["City", "Country", "Max Temp", "Current Description", "Lat", "Lng", "Hotel Name"]]
```

```python
vacation3_df = city_data_df.loc[city_data_df["City"]=="Ebersberg", ["City", "Country", "Max Temp", "Current Description", "Lat", "Lng", "Hotel Name"]]
```

```python
vacation4_df = city_data_df.loc[city_data_df["City"]=="Argentan", ["City", "Country", "Max Temp", "Current Description", "Lat", "Lng", "Hotel Name"]]
```

```python
destination_df=pd.concat([vacation1_df, vacation2_df, vacation3_df, vacation4_df])
destination_df
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""
```

```python
# Store the DataFrame Row.
destination_info = [info_box_template.format(**row) for index, row in destination_df.iterrows()]
for i, r in destination_df.iterrows():
    print(r)
    print(info_box_template.format(**r))
    break
```

```python
locations = destination_df[["Lat", "Lng"]]
# Latitude-longitude pairs
way = [(42.7, -8.25), (48.75, -0.02)]
Isernia = (41.6, 14.24)
Ebersberg = (48.08, 11.97)

fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)

Isernia2Ebersberg_via_SilledaArgentan = gmaps.directions_layer(
        Isernia, Ebersberg, waypoints= way,
        travel_mode='DRIVING')
#fig.add_layer(Isernia2Ebersberg_via_SilledaArgentan)

marker_layer = gmaps.marker_layer(locations, info_box_content=destination_info)
fig.add_layer(marker_layer)


fig
```
